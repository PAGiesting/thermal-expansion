```python
findletters = re.compile('[a-zA-Z]+')
```

```python
finddigits = re.compile('\d+')
finddigits.findall(files[0])
```

```python
dline = {}
pieces = files[0].split('.')
# collect 'c' or 'x' character from end of last piece before extension
dline['Filetype'] = pieces[-2][-1]
stems = pieces[0].split('-')
dline['Sample'] = findletters.findall(stems[0])[0]
dline['Orientation'] = finddigits.findall(stems[0])[0]
prelength = len(dline['Sample'])+len(dline['Orientation'])
dline['Run'] = stems[0][prelength:]
print(pd.DataFrame(dline,index=[0]))
```

```python
for filename in files:
    pieces = filename.split('.')
    dline = {}
    # collect 'c' or 'x' character from end of last piece before extension
    dline['Filetype'] = pieces[-2][-1]
    stems = pieces[0].split('-')
    dline['Sample'] = findletters.findall(stems[0])[0]
    dline['Orientation'] = finddigits.findall(stems[0])[0]
    prelength = len(dline['Sample'])+len(dline['Orientation'])
    dline['Run'] = stems[0][prelength:]
    newline = pd.DataFrame(dline,columns=cols,index=[0])
    namedf = namedf.append(newline,ignore_index=True)
```

```python
def namegrab(text1,text2,df,filelist):
    """Finds the filename in a list that matches text1 and text2
    in columns 1 & 2 of the dataframe and has the lowest value
    for the other columns. Do column 4 first because we trust
    csvs more than workbooks."""
    text4 = df[(df.iloc[:,0]==text1)&(df.iloc[:,1]==text2)].iloc[:,3].min()
    text3 = df[(df.iloc[:,0]==text1)&(df.iloc[:,1]==text2)&(df.iloc[:,3]==text4)].iloc[:,2].min()
    for name in filelist:
        if (text1 in name)&(str(text2) in name)&(str(text3) in name)&(text4 in name):
            return name
    return None
```

```python
plotdict = {}
samples = namedf['Sample'].unique()
for sample in samples:
    plotdict[sample]={}
    or_list = list(namedf[namedf['Sample']==sample].Orientation.unique())
    slots_filled = 0
    if '100' in or_list:
        plotdict[sample]['100']=namegrab(sample,'100',namedf,files)
        slots_filled += 1
        or_list.remove('100')
    if '010' in or_list:
        plotdict[sample]['010']=namegrab(sample,'010',namedf,files)
        slots_filled += 1
        or_list.remove('010')
    if '001' in or_list:
        plotdict[sample]['001']=namegrab(sample,'001',namedf,files)
        slots_filled += 1
        or_list.remove('001')
    while (len(or_list) > 0) & (slots_filled < 3):
        orient = or_list.pop(0)
        if '100' not in plotdict:
            plotdict[sample]['100']=namegrab(sample,orient,namedf,files)
            slots_filled += 1
        elif '010' not in plotdict:
            plotdict[sample]['010']=namegrab(sample,orient,namedf,files)
            slots_filled += 1
        elif '001' not in plotdict:
            plotdict[sample]['001']=namegrab(sample,orient,namedf,files)
            slots_filled += 1
        else:
            print('Error finding an orientation slot.')
```

```python
plotdict
```

```python
for sample in plotdict:
    try:
        df100 = pd.read_csv('data/'+plotdict[sample]['100'])
    except:
        pass
    try:
        df010 = pd.read_csv('data/'+plotdict[sample]['010'])
    except:
        pass
    try:
        df001 = pd.read_csv('data/'+plotdict[sample]['001'])
    except:
        pass
    try:
        plt.plot(df100.iloc[:,1],df100['Alpha'],lw=2,c='r',label='100')
    except:
        pass
    try:
        plt.plot(df010.iloc[:,1],df010['Alpha'],lw=2,c='g',label='010')
    except:
        pass
    try:
        plt.plot(df001.iloc[:,1],df001['Alpha'],lw=2,c='b',label='001')
    except:
        pass
    plt.title(sample)
    plt.xlabel('Temp (Celsius)')
    plt.ylabel('Linear Thermal Exp.')
    plt.legend()
    plt.savefig('plots/'+sample+'.tif')
    plt.show()
    try:
        del df100
    except:
        pass
    try:
        del df010
    except:
        pass
    try:
        del df001
    except:
        pass
```
