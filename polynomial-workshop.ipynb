```python
# A workbook to test the numpy implementation of polynomial fitting
import numpy as np
from numpy.polynomial import Polynomial
import pickle
import openpyxl as xlsx
```

```python
datafile = 'data/sapphire-standard.xlsx'
wb = xlsx.load_workbook(datafile)
```

```python
Tlist = []
dLlist = []
for row in wb.active.values:
    if row[0] == 'T':
        continue
    Tlist.append(float(row[0]))
    dLlist.append(float(row[1])/100.0)
T = np.array(Tlist)
dL = np.array(dLlist)
print(T[0:3])
```

```python
standard = Polynomial.fit(T, dL, 4)
standard
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(T, dL, 'o')
TT, LL = standard.linspace()
plt.plot(TT, LL, lw=1)
print(standard.domain, standard.window)
plt.show()
```

```python
stan5 = Polynomial.fit(T, dL, 5)
plt.plot(T, dL, 'o')
TT, LL = stan5.linspace()
plt.plot(TT, LL, lw=1)
print(stan5.domain, stan5.window)
plt.show()
```

```python
stan6 = Polynomial.fit(T, dL, 6)
plt.plot(T, dL, 'o')
TT, LL = stan6.linspace()
plt.plot(TT, LL, lw=1)
print(stan6.domain, stan6.window)
plt.show()
```

```python
stan3 = Polynomial.fit(T, dL, 3)
plt.plot(T, dL, 'o')
TT, LL = stan3.linspace()
plt.plot(TT, LL, lw=1)
print(stan3.domain, stan3.window)
plt.show()
```

```python
stan2 = Polynomial.fit(T, dL, 2)
plt.plot(T, dL, 'o')
TT, LL = stan2.linspace()
plt.plot(TT, LL, lw=1)
print(stan2.domain, stan2.window)
plt.show()
```

```python
stan2 = Polynomial.fit(T, dL, 2, domain=[0,1500])
plt.plot(T, dL, 'o')
TT, LL = stan2.linspace()
plt.plot(TT, LL, lw=1)
print(stan2.domain, stan2.window)
plt.show()
```

```python
stan4 = Polynomial.fit(T, dL, 4, domain=[0,1250])
plt.plot(T, dL, 'o')
TT, LL = stan4.linspace()
plt.plot(TT, LL, lw=1)
print(stan4.domain, stan4.window)
plt.show()
```

```python
from scipy.interpolate import interp1d
```

```python
stanlin = interp1d(T, dL)
stancub = interp1d(T, dL, kind='cubic')
TT = np.linspace(-200,1700)
plt.plot(T, dL, 'o')
#plt.plot(TT, stanlin(TT), lw=2)
plt.plot(TT, stancub(TT), c='green', lw=2)
plt.xlim(100,300)
plt.show()
```
